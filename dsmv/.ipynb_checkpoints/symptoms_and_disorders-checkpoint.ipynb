```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
import re
import string
```

```python
dsmv = open("dsmv.txt", "r").read()
dsmv = dsmv.replace("Attention- Deficit", "Attention-Deficit")
dsmv = dsmv.replace("attention-deficit /hyperactivity disorder", "attention-deficit/hyperactivity disorder")
dsmv = dsmv.split("\n")

print(len(dsmv))
```

```python
# Remove empty lines and join split lines
for i in reversed(range(len(dsmv))):
    dsmv[i] = dsmv[i].strip()
    if len(dsmv[i]) == 0:
        dsmv.pop(i)
        
    elif len(dsmv[i]) == len([x for x in dsmv[i] if x in string.digits+string.punctuation+string.ascii_letters]):
        dsmv.pop(i)
        
    elif dsmv[i][-1] == "-" or dsmv[i][-1] == "/":
        dsmv[i] = dsmv[i][0:len(dsmv[i])-1]+dsmv[i+1]
        dsmv.pop(i+1)
        
    elif dsmv[i][-1] == ",":
        dsmv[i] = dsmv[i]+" "+dsmv[i+1]
        dsmv.pop(i+1)
        
    elif i < len(dsmv)-1:
        if dsmv[i][-1] != "." and dsmv[i+1][0] not in (string.ascii_uppercase+string.digits):
            dsmv[i] = dsmv[i]+" "+dsmv[i+1]
            dsmv.pop(i+1)

heading_exclusions = [",","."]

for i in reversed(range(len(dsmv))):
    if dsmv[i][-1] == ")" or dsmv[i][-1] in string.digits:
        dsmv.pop(i)
    
    elif len(dsmv[i].split(" ")) == len([1 for x in dsmv[i].split(" ") if x[0] in string.ascii_uppercase]):
        if len([x for x in dsmv[i] if x in heading_exclusions]) > 0:
                dsmv.pop(i)
                
print(len(dsmv))
```

```python
for i in range(len(dsmv)):
    if len(dsmv[i].split(" ")) == len([1 for x in dsmv[i].split(" ") if x[0] in string.ascii_uppercase]):
        if len([x for x in dsmv[i] if x in heading_exclusions]) == 0:
                print(dsmv[i])
                
    elif dsmv[i][0] in (string.ascii_uppercase+string.digits) and dsmv[i][1] == ".":
        print(dsmv[i])
```

```python
occurances = []
verbatim_list = []
verbatim_str = ""
for i in range(len(dsmv)):
    if len(re.findall("Attention-Deficit/Hyperactivity Disorder",dsmv[i]))+len(re.findall("ADHD",dsmv[i])) > 0:
        occurances.append(i)
        verbatim_list.append(dsmv[i])
        
        if len(verbatim_list) > 0:
            verbatim_str = verbatim_str + " " + dsmv[i]
        else:
            verbatim_str = dsmv[i]

print(len(occurances),"lines")
print(len(verbatim_list))
print(verbatim_str)
```

```python
import nltk
import matplotlib.pyplot as plt
```

```python
simple_dist = nltk.FreqDist(verbatim_str.lower().split(" "))
```

```python
simple_dist.plot(cumulative=True)
```
